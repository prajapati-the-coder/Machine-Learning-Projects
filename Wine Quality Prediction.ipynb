Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection


```python
wine_dataset = pd.read_csv('/content/wineQualityReds.csv')
```

```python
wine_dataset.head()
```

```python
wine_dataset.shape
```

```python
# Dropping the Unwanted column
final_columns = ['fixed.acidity', 'volatile.acidity', 'citric.acid', 'residual.sugar', 'chlorides', 'free.sulfur.dioxide', 'total.sulfur.dioxide', 
                 'density', 'pH', 'sulphates', 'alcohol', 'quality']
wine_dataset = wine_dataset.drop(columns = [col for col in wine_dataset if col not in final_columns])
```

```python
wine_dataset.head()
```

```python
wine_dataset.shape
```

```python
#checking for missing values
wine_dataset.isnull().sum()
```

Data Analysis and Visualization

```python
#stastical measures of the dataset
wine_dataset.describe()
```

```python
#number of values for each quality
sns.catplot(x = 'quality', data = wine_dataset, kind = 'count')
```

Data Pre-Processing

```python
# separate the data and label
x = wine_dataset.drop('quality', axis = 1)
# y = wine_dataset['quality']
```

```python
print(x)
```

Label Binarization

```python
y = wine_dataset['quality'].apply(lambda y_value:1 if y_value >=7 else 0)
```

```python
print(y)
```

Train and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state =2, stratify = y)
```

```python
print(x.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

Training the model: Random Forest classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

Evaluation

Accuracy Score

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy of the test data is:', test_data_accuracy)
```

Making a predictive system

```python
input_data = (8.1,0.56,0.28,1.7,0.368,16,56,0.9968,3.11,1.28,9.3)

#changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the data as we are predicting the label for only one instance
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)


prediction = model.predict(input_data_reshape)
print(prediction)

if (prediction[0]==1):
  print("Good Quality Wine")
else:
  print("Bad Quality Wine")
```
